```python
import sqlite3
import pandas as pd

# create database
conn = sqlite3.connect("sales.db")

# sample data
data = {
    "order_id": range(1, 21),
    "customer": ["A","B","C","D","E"] * 4,
    "region": ["North","South","East","West","North"] * 4,
    "sales": [200,150,300,250,400] * 4
}

df = pd.DataFrame(data)

# create table
df.to_sql("sales", conn, if_exists="replace", index=False)

df


query = """
SELECT *
FROM sales
LIMIT 5;
"""

pd.read_sql(query, conn)


query = """
SELECT
    SUM(sales) AS total_sales,
    AVG(sales) AS avg_order_value
FROM sales;
"""

pd.read_sql(query, conn)

query = """
SELECT
    region,
    SUM(sales) AS total_sales
FROM sales
GROUP BY region
ORDER BY total_sales DESC;
"""

pd.read_sql(query, conn)


query = """
SELECT
    customer,
    SUM(sales) AS total_sales
FROM sales
GROUP BY customer
ORDER BY total_sales DESC
LIMIT 5;
"""

pd.read_sql(query, conn)


query = """
SELECT
    customer,
    COUNT(*) AS orders_count,
    SUM(sales) AS total_sales,
    AVG(sales) AS avg_order_value
FROM sales
GROUP BY customer
ORDER BY total_sales DESC;
"""

pd.read_sql(query, conn)


```

## Insights & Recommendations

- Sales performance varies clearly by region, with certain regions generating significantly higher revenue.
- A small number of customers contribute a large portion of total sales.
- Some customers place frequent orders with relatively high average order values.

**Recommendations:**
- Focus retention and loyalty strategies on top-performing customers.
- Increase marketing and sales efforts in high-performing regions.
- Investigate opportunities to grow revenue in lower-performing regions through targeted campaigns.
